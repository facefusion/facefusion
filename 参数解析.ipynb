# 重要参数解析


## processors
参数用于指定要加载或使用的功能模块（处理器）

| 处理器名称 (`processor`) | 功能说明 |
|--------------------------|----------|
| `age_modifier`           | 年龄修改器，对人脸进行年龄变化（可通过 `--age-modifier-model` 和 `--age-modifier-direction` 设置模型和方向）。 |
| `deep_swapper`           | 深度换脸模块，使用预训练模型将源人脸替换到目标视频或图像中。可选择具体模型通过 `--deep-swapper-model`。 |
| `expression_restorer`    | 表情恢复模块，从目标人脸恢复或增强表情，可通过 `--expression-restorer-model` 和 `--expression-restorer-factor` 调整。 |
| `face_debugger`          | 人脸调试模块，可显示人脸检测、关键点、掩码、年龄、性别、种族等调试信息。通过 `--face-debugger-items` 指定显示内容。 |
| `face_editor`            | 人脸编辑模块，可调整眼睛、嘴巴、眉毛、头部姿态等，通过多个 `--face-editor-*` 参数控制细节。 |
| `face_enhancer`          | 人脸增强模块，提高人脸清晰度或修复细节，通过 `--face-enhancer-model` 和 `--face-enhancer-blend` 设置增强模型和融合比例。 |
| `face_swapper`           | 人脸替换模块，用于普通换脸处理（不同于 `deep_swapper` 的深度模型），可选择像素分辨率 `--face-swapper-pixel-boost`。 |
| `frame_colorizer`        | 视频帧上色模块，将黑白或单色帧进行上色处理，通过 `--frame-colorizer-model` 和 `--frame-colorizer-size` 设置模型及帧大小。 |
| `frame_enhancer`         | 视频帧增强模块，提高帧画质，通过 `--frame-enhancer-model` 和 `--frame-enhancer-blend` 设置增强模型和融合比例。 |
| `lip_syncer`             | 唇动同步模块，将音频与目标视频中人脸嘴型对齐，通过 `--lip-syncer-model` 和 `--lip-syncer-weight` 设置模型及权重。 |

###  换脸模块对比

FaceFusion 中有两个换脸处理器：`deep_swapper` 和 `face_swapper`。它们的主要区别如下：

| 参数 | 功能 | 模型特性 | 优点 | 缺点 | 适用场景 |
|------|------|----------|------|------|----------|
| `deep_swapper` | 深度换脸模块 | 基于深度学习（GAN 或深度卷积网络）模型 | 高质量换脸；保留表情、光照、面部细节 | 模型复杂；对 GPU/VRAM 要求高；处理速度较慢 | 高质量换脸、影视级别换脸、复杂表情或角度变化的换脸 |
| `face_swapper` | 普通换脸模块 | 基于人脸对齐和融合的轻量级方法 | 处理速度快；硬件要求低；适合批量或实时处理 | 换脸效果相对普通；细节和光照保真度低 | 快速换脸、实时演示、资源有限的场景、小型视频换脸 |



## 人脸增强
### --face-enhancer-model
faceFusion 的 `face_enhancer` 模块用于**提高人脸图像质量**，修复细节或增强清晰度。可通过参数 `--face-enhancer-model` 选择不同的增强模型。

| 模型名称 | 功能说明 | 适用场景 |
|----------|----------|----------|
| `codeformer` | 高性能人脸修复模型，可在低分辨率或模糊人脸中重建细节，同时保留自然特征 | 模糊或低质量图像的清晰化 |
| `gfpgan_1.2` | 基于 GAN 的人脸修复模型，1.2 版本，专注于保持面部结构自然 | 中等质量图像修复、轻微瑕疵修复 |
| `gfpgan_1.3` | 1.3 版本，相比 1.2 改进了细节保真和肤色一致性 | 高质量人脸增强、保持自然肤色 |
| `gfpgan_1.4` | 最新版本，增强了面部细节还原和抗噪能力 | 高要求人脸增强、复杂光照或噪声场景 |
| `gpen_bfr_256` | GPEN 模型，256x256 分辨率，专注于面部细节和光照修复 | 小分辨率图像的快速增强 |
| `gpen_bfr_512` | GPEN 模型，512x512 分辨率，平衡速度和质量 | 中等分辨率图像增强 |
| `gpen_bfr_1024` | GPEN 模型，1024x1024 分辨率，高精度面部增强 | 高分辨率图像、人像照片增强 |
| `gpen_bfr_2048` | GPEN 模型，2048x2048 分辨率，超高分辨率增强 | 超大尺寸图像、专业修复 |
| `restoreformer_plus_plus` | 最新增强模型，结合 Transformer 技术进行面部修复与细节增强 | 追求最佳画质、人脸细节极致保留 |

#### 使用建议
1. **低分辨率或模糊图像**：使用 `codeformer` 或 `gfpgan_1.4`。
2. **中等分辨率图像**：使用 `gpen_bfr_512` 或 `gfpgan_1.3`。
3. **高分辨率图像**：使用 `gpen_bfr_1024` 或 `gpen_bfr_2048`。
4. **追求最优效果**：尝试 `restoreformer_plus_plus`，需要更多计算资源。
5. **融合比例**：可通过 `--face-enhancer-blend` 设置增强权重（0-100），决定增强效果与原始图像的融合程度。

### --face-enhancer-blend
--face-enhancer-blend` 用于控制 **人脸增强效果与原始图像的融合比例**，值的范围为 0~100。  

| 参数 | 含义 | 取值范围 | 效果说明 | 使用建议 |
|------|------|----------|----------|----------|
| `--face-enhancer-blend` | 控制增强后的脸部与原始图像的融合比例 | 0~100 | - 0：完全保留原始图像，不使用增强效果<br>- 50：原始图像与增强效果各占一半<br>- 100：完全使用增强效果 | - 图像质量较差时，可以设置较高值（70-100）以突出增强效果<br>- 图像本身质量较好时，可设置中等值（30-60）以保持自然感<br>- 追求极致真实感时，可以先尝试 40-60，避免“过度增强”导致不自然 |
| 类型 | 数值 | 整数 | 越高增强越明显 | 结合 `--face-enhancer-weight` 使用，可进一步微调细节强度 |



## 人脸swaper模型
### --face-swapper-model
`--face-swapper-model` 用于选择 **人脸替换模型**，负责将源脸部替换到目标视频或图片上。不同模型在风格、分辨率、速度和精度上有所差异。  

| 模型名称 | 分辨率 | 功能说明 | 适用场景 |
|----------|--------|----------|----------|
| `blendswap_256` | 256x256 | 基础混合替换模型，将源脸与目标脸融合 | 小尺寸图片或快速测试 |
| `ghost_1_256` | 256x256 | 高精度面部融合，减少边缘痕迹 | 中小尺寸人脸视频 |
| `ghost_2_256` | 256x256 | 改进版 Ghost 模型，增强光照一致性 | 中小尺寸视频，保持光照自然 |
| `ghost_3_256` | 256x256 | Ghost 系列最稳版本，减少瑕疵和失真 | 视频或图像质量要求较高 |
| `hififace_unofficial_256` | 256x256 | HiFiFace 模型非官方版本，增强细节 | 面部细节重建、表情保留 |
| `hyperswap_1a_256` | 256x256 | HyperSwap 系列，快速人脸替换 | 需要快速渲染的场景 |
| `hyperswap_1b_256` | 256x256 | HyperSwap 改进版，增强光照与肤色自然度 | 视频中连续帧替换 |
| `hyperswap_1c_256` | 256x256 | HyperSwap 稳定版，减少瑕疵和错位 | 高质量视频替换 |
| `inswapper_128` | 128x128 | 小尺寸人脸替换，速度快 | 快速测试、小头像替换 |
| `inswapper_128_fp16` | 128x128 | FP16 加速版本，减少显存占用 | GPU 显存有限场景 |
| `simswap_256` | 256x256 | SimSwap 模型，注重保留表情 | 中小尺寸视频，强调表情传递 |
| `simswap_unofficial_512` | 512x512 | 非官方高分辨率 SimSwap，保留更多细节 | 高分辨率图像或视频 |
| `uniface_256` | 256x256 | 通用人脸替换模型 | 小到中尺寸人脸图像 |

#### 使用建议
1. **快速测试或小图像**：使用 `inswapper_128` 或 `blendswap_256`。  
2. **中等分辨率视频**：使用 `ghost_2_256`、`simswap_256` 或 `hyperswap_1b_256`。  
3. **高质量视频或图像**：使用 `simswap_unofficial_512` 或 `hyperswap_1c_256`。  
4. **追求细节与自然感**：优先选择 Ghost 系列或 HyperSwap 稳定版。  

#### 注意
- 可配合 `--face-swapper-pixel-boost` 调整输出分辨率，进一步提高人脸细节。  
- 不同模型对表情、光照和肤色处理效果不同，可根据目标效果选择。

## --face-swapper-pixel-boost
`--face-swapper-pixel-boost` 用于设置 **人脸替换输出的像素增强分辨率**，即在执行 face swap 时，对结果人脸进行放大或细化处理，以提升细节和清晰度。

| 参数 | 说明 | 适用场景 | 建议值 |
|------|------|----------|---------|
| `128x128` | 输出人脸较小，速度快，占用显存低 | 小图像、人脸头像替换、快速测试 | 测试或显存有限场景 |
| `256x256` | 默认中等分辨率，平衡速度和效果 | 常规视频或图片替换 | 推荐多数场景 |
| `384x384` | 提升细节，轻微增加显存占用 | 中等分辨率视频，需要细节更好 | 表情细节较重要时使用 |
| `512x512` | 高分辨率人脸替换，显存占用较高 | 高质量图像或视频 | 输出质量要求高时使用 |
| `768x768` | 超高分辨率，显存占用大 | 高清视频或静态图片处理 | 高精度需求，GPU 显存充足时 |
| `1024x1024` | 极致清晰度，显存占用非常高 | 超高清图像或特效需求 | 高端 GPU 场景下使用 |

#### 使用说明
- 结合 `--face-swapper-model` 使用，不同模型对高分辨率支持程度不同。  
- 提高 `pixel-boost` 分辨率可增强面部细节，但会增加计算时间和显存消耗。  
- 对于连续视频处理，建议选择中等分辨率（256~384），避免 GPU 内存不足或处理速度过慢。  


## 人脸挑选模型
### --face-selector-mode
`--face-selector-mode` 用于选择 **如何在多个人脸中挑选目标人脸** 的策略，即在一张图片或视频帧中存在多张人脸时，程序选择哪一张作为处理对象。

| 参数 | 说明 | 使用场景 | 建议 |
|------|------|----------|------|
| `many` | 保留图像或视频中 **所有检测到的人脸**，对每张人脸都执行后续处理 | 多人群像、群体处理、多人替换 | 当需要对每张人脸都做处理时使用 |
| `one` | 只选择 **一张人脸**，通常根据顺序或位置（可配合 `--face-selector-order`） | 单人肖像或视频，忽略其他人脸 | 视频中只有目标人物时使用，减少计算量 |
| `reference` | 根据 **参考人脸匹配** 来选择最相似的人脸 | 使用 reference face 或模板人脸做匹配 | 当需要精确匹配目标人物时使用，常用于深度换脸场景 |

### 其他相关参数
- `--face-selector-order`：当 `one` 模式下可指定选择顺序，如 `left-right`、`top-bottom` 等。  
- `--reference-face-position` 和 `--reference-face-distance`：当使用 `reference` 模式时，用于指定参考人脸位置和匹配相似度。  
- `--face-selector-age-start` / `--face-selector-age-end` / `--face-selector-gender` / `--face-selector-race`：可进一步筛选符合条件的人脸。

## 人脸检测模型
### --face-detector-model
`--face-detector-model` 用于选择 **用于检测人脸的模型**，即在输入图像或视频中识别出人脸的位置和数量。不同模型在精度、速度、GPU占用和检测角度支持上有所差异。

| 参数 | 说明 | 特点 | 使用场景 |
|------|------|------|----------|
| `many` | 自动选择或组合多模型检测人脸 | 综合精度与速度，默认模式 | 适合大多数场景，适应性强 |
| `retinaface` | 高精度人脸检测，适合各种角度 | 精度高，速度适中 | 对精细人脸识别要求高的静态图片或视频 |
| `scrfd` | 超轻量级、高速度人脸检测 | 速度快，GPU消耗低 | 视频流或实时处理场景 |
| `yolo_face` | 基于 YOLO 的人脸检测 | 速度快，适合大尺寸帧 | 视频或多目标场景，适合快速推理 |

#### 相关参数
- `--face-detector-size`：指定检测器处理的图像分辨率（如 `640x640`），分辨率越高精度越高但速度越慢。  
- `--face-detector-angles`：可指定旋转角度，检测不同朝向的人脸。  
- `--face-detector-score`：过滤低置信度人脸，取值范围 `[0.0..1.0]`。

### 使用示例
```bash
# 使用 RetinaFace 检测人脸
python facefusion.py run \
  --face-detector-model retinaface \
  --face-detector-size 640x640 \
  --face-detector-score 0.8

# 使用 SCRFD 高速检测
python facefusion.py run \
  --face-detector-model scrfd \
  --face-detector-size 640x640
```

## 面部关键点检测模型
### --face-landmarker-model
`--face-landmarker-model` 用于选择面部关键点检测模型（Face Landmarker），主要负责检测面部的关键点位置（眼睛、嘴巴、鼻子、眉毛等），为换脸、表情修复、面部编辑等提供精准定位。

| 参数 | 说明 | 可选值 | 特点 | 使用场景 |
|------|------|--------|------|----------|
| `--face-landmarker-model` | 面部关键点检测模型选择 | `many` / `2dfan4` / `peppa_wutz` | - `many`: 检测多个关键点（精度高、速度中等）<br>- `2dfan4`: 基于 2D FAN 网络，检测精度较高，轻量化<br>- `peppa_wutz`: 特定优化模型，速度快，适合低显存环境 | - 高质量换脸和表情修复，建议用 `many` 或 `2dfan4`<br>- 实时处理或低显存场景可使用 `peppa_wutz` |

#### 模型特点说明
1. **many**
   - 检测较多关键点（可达到 68 点或更多）

2. **2dfan4**
   - 基于 2D FAN（Face Alignment Network）网络
   - 精度高，但模型较轻量

   - 优化轻量化模型
   - 检测速度快
   - 精度略低，适合实时处理或低显卡环境

## 人脸遮挡区域检测模型

| 参数 | 说明 | 可选值 / 范围 | 特点 | 使用场景 |
|------|------|---------------|------|----------|
| `--face-occluder-model` | 人脸遮挡区域检测模型 | `xseg_1` / `xseg_2` / `xseg_3` | 检测口罩、眼镜、帽子等遮挡区域 | 处理多人视频或复杂遮挡场景 |
| `--face-parser-model` | 面部区域分割模型 | `bisenet_resnet_18` / `bisenet_resnet_34` | 生成精细的面部区域掩码 | 高质量换脸、增强、编辑 |
| `--face-mask-types` | 掩码类型 | `box` / `occlusion` / `area` / `region` | 决定掩码生成的精细程度 | 结合 occluder/parser 使用，覆盖不同区域 |
| `--face-mask-areas` | 面部区域掩码选择 | `upper-face` / `lower-face` / `mouth` | 与 `area` 类型配合使用 | 仅处理特定面部区域 |
| `--face-mask-regions` | 面部细分区域掩码 | `skin`, `left-eyebrow`, `right-eyebrow`, `left-eye`, `right-eye`, `glasses`, `nose`, `mouth`, `upper-lip`, `lower-lip` | 与 `region` 类型配合使用 | 精细化控制面部区域掩码 |
| `--face-mask-blur` | 掩码边缘模糊程度 | `[0.0..1.0:0.05]` | 提升融合效果，减少边缘突兀感 | 高质量换脸或面部增强 |
| `--face-mask-padding` | 掩码区域扩展 | `[top right bottom left]` 数值 | 扩展掩码边界，防止处理遗漏 | 避免换脸或增强时裁切到面部边缘 |


### --face-occluder-model
`--face-occluder-model` 用于选择 **人脸遮挡区域检测模型**，即识别图像或视频中人脸被遮挡的部分（如口罩、眼镜、帽子等），以便在后续处理（如换脸或增强）中正确处理遮挡区域。

| 参数 | 说明 | 特点 | 使用场景 |
|------|------|------|----------|
| `xseg_1` | 人脸遮挡检测模型版本 1 | 精度较高，适合中等分辨率图像 | 普通视频或图片处理 |
| `xseg_2` | 人脸遮挡检测模型版本 2 | 精度更高，计算量适中 | 对遮挡细节要求较高的场景 |
| `xseg_3` | 人脸遮挡检测模型版本 3 | 最新版本，精度和鲁棒性最佳 | 高质量视频换脸或复杂遮挡场景 |

#### 使用示例
```bash
# 使用 xseg_1 模型
python facefusion.py run \
  --face-occluder-model xseg_1

# 使用最新的 xseg_3 模型处理复杂遮挡
python facefusion.py run \
  --face-occluder-model xseg_3
```

### --face-mask-types
`--face-mask-types` 用于选择 **生成的人脸掩码类型**，即在处理换脸、增强或编辑时指定哪些部分需要掩盖或处理。可以同时选择多种类型进行组合。

| 参数 | 说明 | 特点 | 使用场景 |
|------|------|------|----------|
| `box` | 使用简单矩形框掩盖人脸区域 | 速度快，简单粗略 | 对快速处理要求高的场景 |
| `occlusion` | 遮挡掩码，检测口罩、眼镜、帽子等遮挡部分 | 精度较高，可处理复杂遮挡 | 视频或多人场景下的遮挡处理 |
| `area` | 面部区域掩码，通常覆盖上脸、下脸、嘴巴等 | 可自定义处理区域 | 对特定面部区域增强或换脸 |
| `region` | 精细区域掩码，覆盖皮肤、眼睛、眉毛、鼻子、嘴巴等 | 精度最高，可细粒度操作 | 高质量换脸或面部编辑场景 |

#### 使用示例
```bash
# 使用 box 和 occlusion 掩码类型
python facefusion.py run \
  --face-mask-types box occlusion

# 使用精细区域掩码
python facefusion.py run \
  --face-mask-types region
```

### --face-parser-model
`--face-parser-model` 用于选择面部分割模型（Face Parser），主要负责将面部划分成不同区域（皮肤、五官、口唇等），生成精细的面部掩码，以便换脸、增强或编辑等处理。

| 参数 | 说明 | 可选值 | 特点 | 使用场景 |
|------|------|--------|------|----------|
| `--face-parser-model` | 面部分割模型选择 | `bisenet_resnet_18` / `bisenet_resnet_34` | - 输出精细的面部区域掩码<br>- `bisenet_resnet_34` 精度更高，但计算开销更大<br>- `bisenet_resnet_18` 精度稍低，但速度更快 | - 高质量换脸时，建议用 `bisenet_resnet_34`<br>- 实时或低显卡场景可以选择 `bisenet_resnet_18` |

### 模型特点说明
1. **bisenet_resnet_18**
   - 基于 BiSeNet 和 ResNet-18
   - 模型轻量，速度快
   - 面部区域分割精度中等
   - 适合实时视频处理或显存有限的设备

2. **bisenet_resnet_34**
   - 基于 BiSeNet 和 ResNet-34
   - 模型更大，计算开销高
   - 面部区域分割精度高，尤其对眉毛、眼睛、嘴唇等细节表现更好
   - 适合高质量换脸、增强和精细编辑任务

### 使用示例

```bash
# 使用高精度分割模型
python facefusion.py run \
  --face-parser-model bisenet_resnet_34 \
  --face-mask-types region \
  --face-mask-regions skin left-eye right-eye nose mouth
